```python
# import the datareader
import pandas_datareader.data as web
# import the date conversion function
import datetime
```

```python
# initialize the start date
start = datetime.datetime(2010, 2, 1)
# initialize the end date
end = datetime.datetime(2013, 1, 27)
# get the IBM stock data for a given period from Yahoo Finance
data = web.DataReader("IBM", 'yahoo',start,end)
# show the first 5 observations
data.head()
```

```python
# the following command will save the data for future usage
# please, note, it will save as a ".csv" file, but one can save in other formats also
data.to_csv("IBM_data.csv")
```

```python
# import the plotting library
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# plot the opening and closing stock prices for 2012 only
plt.plot(data['Open']['2012'])
plt.plot(data['Close']['2012'])

# the following command will save the plot as a .png (image) file
plt.savefig("IBM_plot.png")
```
