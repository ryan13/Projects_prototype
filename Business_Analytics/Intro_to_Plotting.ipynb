# Introduction to Matplotlib (by Hovhannes Mkrtchyan)

Matplotlib is a plotting library for Python which makes both easy and hard thing possible. This notebook very shortly introduces some key properties of the library.

```python
# Import the library
import matplotlib.pyplot as plt 
```

```python
# Create a plot on a sample data
plt.plot([1,2,3,4])
plt.show()
```

```python
# Provide values for both X and Y axis
plt.plot([1,2,3,4], [5,6,7,8])
plt.show()
```

```python
plt.plot([1,2,3,4], [5,6,7,8])
# "zoom out" the plot by changing the values of axis
plt.axis([0,10,0,20])
# provide label for Y axis
plt.ylabel('Dollars')
# provide label for X axis
plt.xlabel('Years')
plt.show()
```

```python
# Change the presentation of the plot to RED and Circled (ro)
plt.plot ([1,2,3,4], [5,6,7,8], 'ro')
# In the same way one could change it to green trangles by using 'g^',
# to blue dashes by using 'b--' commands and etc.
plt.axis([0,10,0,15])
plt.ylabel('Dollars')
plt.xlabel('Years')
# provide title for the plot
plt.title('Business Analytics')
plt.show()
```
